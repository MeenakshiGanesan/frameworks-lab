```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python

df = pd.read_csv("/content/insurance.csv")
print(df.head())
```

```python
plt.figure(figsize=(8,5))
plt.plot(df['age'], df['charges'], 'o-', color='green')
plt.title("Line Chart: Age vs Insurance Charges")
plt.xlabel("Age")
plt.ylabel("Charges")
plt.show()
```

```python
avg_charges = df.groupby('smoker')['charges'].mean()

plt.figure(figsize=(6,5))
avg_charges.plot(kind='bar', color=['skyblue','salmon'])
plt.title("Bar Chart: Average Charges by Smoker Status")
plt.ylabel("Average Charges")
plt.xlabel("Smoker")
plt.show()
```

```python
plt.figure(figsize=(7,5))
plt.hist(df['bmi'], bins=20, color='purple', edgecolor='black')
plt.title("Histogram: Distribution of BMI")
plt.xlabel("BMI")
plt.ylabel("Frequency")
plt.show()
```
